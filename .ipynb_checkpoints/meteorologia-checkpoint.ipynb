```python
from sqlalchemy import create_engine
import pandas as pd

# Defina suas credenciais
usuario = 'root'
senha = '3101'
host = 'localhost'
banco_de_dados = 'meteorologia'

# Criar uma conexão com o MySQL
engine = create_engine(f'mysql+mysqlconnector://{usuario}:{senha}@{host}/{banco_de_dados}')

# Preparar os dados para serem inseridos na tabela
dados_limpos_copy = dados_limpos.copy()  # Cria uma cópia do DataFrame
dados_limpos_copy.rename(columns={
    'DATE': 'data', 
    'TAVG': 'temperatura', 
    'PRCP': 'precipitacao'
}, inplace=True)

# Inserir os dados na tabela
dados_limpos_copy[['data', 'temperatura', 'precipitacao']].to_sql('dados_meteorologicos', con=engine, if_exists='append', index=False)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sqlalchemy import create_engine

# Estabelecer a conexão com o banco de dados
engine = create_engine('mysql+pymysql://root:3101@localhost/meteorologia')

# Carregar dados da tabela 'dados_meteorologicos' de volta para um DataFrame
dados_sql = pd.read_sql('SELECT * FROM dados_meteorologicos', con=engine)

# Garantir que a coluna 'data' está no formato datetime
dados_sql['data'] = pd.to_datetime(dados_sql['data'])

# Agrupar dados por mês e calcular a média de precipitação
dados_sql['mes'] = dados_sql['data'].dt.to_period('M')
media_precipitacao = dados_sql.groupby('mes')['precipitacao'].mean()

# Plotar
plt.figure(figsize=(10, 5))
media_precipitacao.plot(kind='bar', color='skyblue', edgecolor='black')
plt.title('Média de Precipitação por Mês')
plt.xlabel('Mês')
plt.ylabel('Precipitação Média (mm)')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()
```
